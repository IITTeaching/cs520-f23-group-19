```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
df = pd.read_csv('./cleaned.csv')
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# visualization of budget vs rating
sns.scatterplot(x='budget', y='vote_average', data=df)

# graph label
plt.xlabel('Budget')
plt.ylabel('Rating')
plt.title('Budget vs Rating')

plt.show()
```

```python
genres = df.columns[12:32]
genres
```

```python
# genre distribution
sns.barplot(x=df[genres].sum().index, y = df[genres].sum().values)

plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Genres')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# distribution by decade
df['year'] = pd.to_datetime(df['release_date']).dt.year
df['decade'] = (df['year'] // 10) * 10
by_decade = df['decade'].value_counts().sort_index()

sns.barplot(x=by_decade.index, y=by_decade.values)

plt.xlabel('Decade')
plt.ylabel('# of Releases')
plt.title('Number of Releases by Decade')

plt.xticks(rotation=45, ha='right')

plt.show()
```
